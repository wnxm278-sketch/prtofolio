```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
folder_path = "Airbnb_project"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
print(csv_files)
```

```python
dataframes = []
for file in csv_files:
    filename = os.path.basename(file).replace(".csv", "").lower()

    if "weekdays" not in filename and "weekends" not in filename:
        continue

    df = pd.read_csv(file)

    parts = filename.split("_")

    city = parts[0]
    day_type = parts[1]

    df["City"] = city.capitalize()
    df["DayType"] = day_type.capitalize()

    dataframes.append(df)

airbnb_df = pd.concat(dataframes, ignore_index=True)
```
